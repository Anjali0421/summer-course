```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```
df = pd.read_csv("cleaned_mall_customers.csv")
df.shape
```

```
df
```

```
df.describe()
```

```
# Create histogram for age distribution
plt.figure(figsize=(6, 4))
plt.hist(df['Age'],color='red',edgecolor= 'black',bins=5,alpha=0.8)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_histogram.png')
plt.show()
```

```
# Create histogram for annual income distribution
plt.figure(figsize=(6, 4))
plt.hist(df['Annual Income'],color='green',edgecolor= 'black',bins=5,alpha=0.8)
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.savefig('annual_income_histogram.png')
plt.show()
```

```
# Create histogram for spending score distribution
plt.figure(figsize=(6, 4))
plt.hist(df['Spending Score'],color='pink',edgecolor= 'black',bins=5,alpha=0.8)
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.savefig('spending_score_histogram.png')
plt.show()
```

```
# Create a scatter plot of annual income vs. spending score, with the data points colored by gender
plt.figure(figsize=(6, 4))
plt.scatter(df['Annual Income'], df['Spending Score'], c=df['Gender'], cmap='viridis')
plt.title('Annual Income vs. Spending Score by Gender')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.colorbar(label='Gender (0 = Male, 1 = Female)')
plt.savefig('EDA_Scatter.png')
plt.show()
```
