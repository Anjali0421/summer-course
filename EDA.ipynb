```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```
df = pd.read_csv("cleaned_mall_customers.csv")
df.shape
```

```
# Calculate descriptive statistics
print("Descriptive Statistics:")
print(df.describe())
```

```
# Correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='YlOrRd')
plt.title('Correlation Matrix')
plt.show()
```

```
# Create histogram for age distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=8, edgecolor= 'black', color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('number of customers')
plt.savefig('age_histogram.png')
plt.show()
```

```
# Create histogram for annual income distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Annual Income'], bins=8, edgecolor= 'black', color='pink')
plt.title('Annual Income Distribution')
plt.xlabel('Annual Income in k$')
plt.ylabel('Frequency')
plt.savefig('annual_income_histogram.png')
plt.show()
```

```
# Create histogram for spending score distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Spending Score'],color='yellow',bins=8, edgecolor= 'black')
plt.title('Spending Score Distribution')
plt.xlabel('Spending Score')
plt.ylabel('Frequency')
plt.savefig('spending_score_histogram.png')
plt.show()
```

```
# Calculate the average spending score for each gender
spending_by_gender = df.groupby('Gender')['Spending Score'].mean()

# Create the bar chart
plt.figure(figsize=(5, 4))
spending_by_gender.plot(kind='bar', color='green')
plt.title('Average Spending Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Score')
plt.show()
```

```
# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Annual Income'], label='Annual Income')
plt.plot(df['Spending Score'], label='Spending Score')
plt.title('Annual Income vs Spending Score')
plt.xlabel('Customer')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```
# Scatter plot of annual income vs spending score
plt.figure(figsize=(6, 4))
plt.scatter(df['Annual Income'], df['Spending Score'])
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```
# Calculate the average spending score for each gender
spending_by_gender = df.groupby('Gender')['Spending Score'].mean()

# Create the bar chart
plt.figure(figsize=(5, 4))
spending_by_gender.plot(kind='bar', color='green')
plt.title('Average Spending Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Spending Score')
plt.show()
```
